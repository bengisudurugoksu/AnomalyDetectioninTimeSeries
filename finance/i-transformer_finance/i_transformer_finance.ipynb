```python
# ============================================================
# 1) IMPORTS
# ============================================================
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import register_keras_serializable
import matplotlib.pyplot as plt

print("TF version:", tf.__version__)
print("tf.keras version:", tf.keras.__version__)

# ============================================================
# 2) GOOGLE DRIVE MOUNT
# ============================================================
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# ============================================================
# 3) LOAD PREPARED FINANCE DATA (64,9)
# ============================================================
X = np.load("/content/drive/MyDrive/anomaly_project/data/windowed/X_finance_windowed_v3.npy")
y = np.load("/content/drive/MyDrive/anomaly_project/data/windowed/y_finance_windowed_v3.npy")

print("Loaded:", X.shape, y.shape)

# Train ONLY on normal windows
X_train = X[y == 0]
print("Training shape:", X_train.shape)

# ============================================================
# 4) SAFE SERIALIZABLE iTransformer BLOCK
# ============================================================
@register_keras_serializable()
class iTransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1, **kwargs):
        super().__init__(**kwargs)
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.ff_dim = ff_dim
        self.rate = rate

        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim, activation="relu"),
            layers.Dense(embed_dim),
        ])
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training=None):
        attn_output = self.att(inputs, inputs, training=training)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)

        ffn_output = self.ffn(out1, training=training)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

    def get_config(self):
        config = super().get_config()
        config.update({
            "embed_dim": self.embed_dim,
            "num_heads": self.num_heads,
            "ff_dim": self.ff_dim,
            "rate": self.rate,
        })
        return config

# ============================================================
# 5) BUILD iTRANSFORMER AUTOENCODER
# ============================================================
def build_iTransformer_AE(seq_len=64, num_features=9, embed_dim=64, num_heads=4, ff_dim=128, num_layers=2):

    inputs = layers.Input(shape=(seq_len, num_features))
    x = layers.Dense(embed_dim)(inputs)

    for _ in range(num_layers):
        x = iTransformerBlock(embed_dim, num_heads, ff_dim)(x)

    outputs = layers.Dense(num_features)(x)

    model = Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss="mse"
    )
    return model

# Build model
itransformer_model = build_iTransformer_AE()
itransformer_model.summary()

# ============================================================
# 6) CALLBACKS — EN SAĞLAM FORMAT
# ============================================================
checkpoint_savedmodel = "/content/drive/MyDrive/anomaly_project/models/itransformer_finance_best"
checkpoint_h5 = "/content/drive/MyDrive/anomaly_project/models/itransformer_finance_best.h5"

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True,
    mode='min'
)

checkpoint_h5_cb = ModelCheckpoint(
    filepath=checkpoint_h5,
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    save_weights_only=False
)

lr_reduce = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)



# ============================================================
# 7) TRAIN
# ============================================================
history = itransformer_model.fit(
    X_train, X_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, checkpoint_h5_cb, lr_reduce],
    verbose=1
)



# ============================================================
# 8) PLOT TRAINING CURVE
# ============================================================
plt.figure(figsize=(7,4))
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```
